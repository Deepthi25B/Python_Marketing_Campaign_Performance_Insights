# **Python DA Capstone Project - Marketing Campaign Performance Insights**

```python
#1) Load Dataset
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
# Load dataset
url = "https://raw.githubusercontent.com/ArchanaInsights/Datasets/main/marketing_campaign.csv"
df = pd.read_csv(url)
# Convert Date column
df['Date'] = pd.to_datetime(df['Date'], format='%d-%m-%Y')
print("Dataset Loaded Successfully")

#2) Descriptive Analysis
#Basic Structure
print(df.head())
print("Rows and Columns:", df.shape)
print(df.info())
print(df.describe())
#Data Exploration
print("Unique Campaign_ID:", df['Campaign_ID'].nunique())
print("Locations:", df['Location'].unique())
print("Customer Segments:", df['Customer_Segment'].unique())
print(df['Campaign_Type'].value_counts())
print(df['Channel_Used'].value_counts())
#3) Exploratory Data Analysis (EDA)
sns.set(style="whitegrid")
plt.figure(figsize=(8,5))
#Campaign Performance
#Scatter: Acquisition Cost vs ROI
plt.figure(figsize=(8,5))
sns.scatterplot(data=df, x="Acquisition_Cost", y="ROI", hue="Campaign_Type")
plt.title("Acquisition Cost vs ROI")
plt.show()
#Avg Conversion Rate by Channel & Campaign Type
plt.figure(figsize=(10,6))
sns.barplot(data=df, x="Channel_Used", y="Conversion_Rate", hue="Campaign_Type")
plt.xticks(rotation=45)
plt.title("Avg Conversion Rate by Channel & Campaign Type")
plt.show()
#Engagement Score Distribution (Box)
plt.figure(figsize=(8,5))
sns.boxplot(data=df, x="Campaign_Type", y="Engagement_Score")
plt.title("Engagement Score by Campaign Type")
plt.show()
#Avg ROI by Company
plt.figure(figsize=(10,5))
df.groupby("Company")["ROI"].mean().plot(kind="bar")
plt.title("Average ROI by Company")
plt.show()
#Heatmap (Engagement vs Conversion)
plt.figure(figsize=(6,4))
sns.heatmap(df[["Engagement_Score","Conversion_Rate"]].corr(), annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.show()
#Customer Segmentation
#Target Audience Distribution

plt.figure(figsize=(10,5))
sns.countplot(data=df, x="Target_Audience")
plt.xticks(rotation=45)
plt.title("Target Audience Distribution")
plt.show()
#Highest Conversion Rate Segment per Language
plt.figure(figsize=(10,6))
sns.barplot(data=df, x="Language", y="Conversion_Rate", hue="Customer_Segment")
plt.title("Conversion Rate by Language & Customer Segment")
plt.show()
#Acquisition Cost Distribution by Segment & Channel
plt.figure(figsize=(10,6))
sns.boxplot(data=df, x="Customer_Segment", y="Acquisition_Cost", hue="Channel_Used")
plt.xticks(rotation=45)
plt.title("Acquisition Cost Distribution")
plt.show()
#Avg Conversion Rate by Language
plt.figure(figsize=(8,5))
df.groupby("Language")["Conversion_Rate"].mean().plot(kind="bar")
plt.title("Avg Conversion Rate by Language")
plt.show()
#Channel Effectiveness
#Engagement Score by Channel & Campaign
plt.figure(figsize=(10,6))
sns.barplot(data=df, x="Channel_Used", y="Engagement_Score", hue="Campaign_Type")
plt.xticks(rotation=45)
plt.title("Engagement Score by Channel & Campaign")
plt.show()
#ROI Distribution by Channel (Pie)
roi_channel = df.groupby("Channel_Used")["ROI"].sum()

plt.figure(figsize=(6,6))
plt.pie(roi_channel, labels=roi_channel.index, autopct="%1.1f%%")
plt.title("ROI Distribution by Channel")
plt.show()

#Clicks vs Impressions
plt.figure(figsize=(8,5))
sns.scatterplot(data=df, x="Clicks", y="Impressions", hue="Channel_Used")
plt.title("Clicks vs Impressions")
plt.show()

#Time-Based Analysis
#Duration Histogram
plt.figure(figsize=(8,5))
plt.hist(df["Duration"], bins=20)
plt.title("Duration Distribution")
plt.show()
#Conversion Rate Over Time by Company
plt.figure(figsize=(12,6))
sns.lineplot(data=df, x="Date", y="Conversion_Rate", hue="Company")
plt.title("Conversion Rate Over Time")
plt.show()
#Engagement Trend Over Time
plt.figure(figsize=(12,6))
sns.lineplot(data=df, x="Date", y="Engagement_Score")
plt.title("Engagement Score Trend")
plt.show()
#Geographic Analysis
#Highest Acquisition Cost by Location
plt.figure(figsize=(10,5))
df.groupby("Location")["Acquisition_Cost"].mean().plot(kind="bar")
plt.title("Acquisition Cost by Location")
plt.show()
#Conversion Rate by Location & Audience
plt.figure(figsize=(12,6))
sns.barplot(data=df, x="Location", y="Conversion_Rate", hue="Target_Audience")
plt.xticks(rotation=45)
plt.title("Conversion Rate by Location & Audience")
plt.show()
#ROI Proportion by Location (Pie)
roi_loc = df.groupby("Location")["ROI"].sum()

plt.figure(figsize=(6,6))
plt.pie(roi_loc, labels=roi_loc.index, autopct="%1.1f%%")
plt.title("ROI by Location")
plt.show()
```
